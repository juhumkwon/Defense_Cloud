```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 무작위 데이터 생성 (200개 샘플, 2차원)
"""
n_samples=200 → 200개의 샘플을 생성
centers=3 → 3개의 군집(클러스터) 중심을 생성
cluster_std=1.0 → 각 군집의 표준 편차(클러스터 퍼짐 정도) 설정
random_state=42 → 재현성을 위해 난수 시드를 고정
X → 생성된 데이터 포인트 (shape: (200, 2))
"""
X, _ = make_blobs(n_samples=200, centers=3, cluster_std=1.0, random_state=42)

# K-평균 군집화 모델 생성 (클러스터 개수: 3)
"""
n_clusters=3 → 3개의 클러스터로 데이터를 분류
random_state=42 → 결과 재현성을 위해 난수 시드 고정
n_init=10 → 초기 중심값을 다르게 설정하여 10번 반복 후 최적의 결과 선택
"""
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

# 군집 중심 좌표
"""
kmeans.cluster_centers_ → 학습된 클러스터 중심 좌표 (shape: (n_clusters, n_features))
centroids 변수에 저장하여 나중에 활용 가능
"""
centroids = kmeans.cluster_centers_

# 예측된 클러스터 레이블
labels = kmeans.labels_

# 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', alpha=0.7)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
