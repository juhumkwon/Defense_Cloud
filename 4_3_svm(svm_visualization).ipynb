```python
from sklearn import datasets
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

# 데이터셋 로드 (아이리스 데이터)
iris = datasets.load_iris()

# 특성 3 (petal length)와 특성 4 (petal width) 사용
X = iris.data[:, 2:4]  # 특성 3, 4
y = iris.target  # 정답 레이블

# SVM 모델 생성 (RBF 커널 사용)
model = svm.SVC(kernel='rbf', gamma='scale')

# 모델 학습
model.fit(X, y)

# 2D 데이터에 대한 결정 경계를 시각화
xx, yy = np.meshgrid(np.linspace(0, 7, 100), np.linspace(0, 3, 100))  # 특성 3과 4의 범위에 맞는 그리드
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 시각화
plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='coolwarm', s=100)
plt.title("SVM with petal length and petal width")
plt.xlabel("Petal Length (Feature 3)")
plt.ylabel("Petal Width (Feature 4)")
plt.show()
```
