```python
from sklearn import datasets
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt

#   ( )
iris = datasets.load_iris()

#  3 (petal length)  4 (petal width) 
X = iris.data[:, 2:4]  #  3, 4
y = iris.target  #  

# SVM   (RBF  )
model = svm.SVC(kernel='rbf', gamma='scale')

#  
model.fit(X, y)

# 2D     
xx, yy = np.meshgrid(np.linspace(0, 7, 100), np.linspace(0, 3, 100))  #  3 4   
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 
plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='coolwarm', s=100)
plt.title("SVM with petal length and petal width")
plt.xlabel("Petal Length (Feature 3)")
plt.ylabel("Petal Width (Feature 4)")
plt.show()
```
